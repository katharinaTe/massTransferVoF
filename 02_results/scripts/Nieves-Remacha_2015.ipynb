```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

Reference: Nieves-Remacha, M. J., Yang, L., & Jensen, K. F. (2015). OpenFOAM computational fluid dynamic simulations of two-phase flow and mass transfer in an advanced-flow reactor. Industrial & Engineering Chemistry Research, 54(26), 6649-6659.

# Nieves-Remacha Figure SI-2

```python
ref_si2= pd.read_csv('../data/Nieves-Remacha_2015/Nieves_Remacha_SI-2.csv', header=None, names=['x','y'])
cfs_si2_001= pd.read_csv('../data/Nieves-Remacha_2015/Nieves_Remacha_SI2_001s.csv')
cfs_si2_005= pd.read_csv('../data/Nieves-Remacha_2015/Nieves_Remacha_SI2_005s.csv')
cfs_si2_010= pd.read_csv('../data/Nieves-Remacha_2015/Nieves_Remacha_SI2_010s.csv')
cfs_si2_020= pd.read_csv('../data/Nieves-Remacha_2015/Nieves_Remacha_SI2_020s.csv')
cfs_si2_030= pd.read_csv('../data/Nieves-Remacha_2015/Nieves_Remacha_SI2_030s.csv')
cfs_si2_040= pd.read_csv('../data/Nieves-Remacha_2015/Nieves_Remacha_SI2_040s.csv')
cfs_si2_100= pd.read_csv('../data/Nieves-Remacha_2015/Nieves_Remacha_SI2_100s.csv')
```

```python
# raw data
fig = plt.figure(figsize=(15, 10))
ax = fig.add_subplot()
ax.plot(ref_si2['x'], ref_si2['y'], marker ='o', ms=3 ,label='Reference, Nieves-Remacha et al. (2015)', linestyle='None', color='#543005')
ax.plot(cfs_si2_001['Points:1'], cfs_si2_001['C'], linestyle='-',label='massTransferVoF, t = 01 s', color='#8c510a')
ax.plot(cfs_si2_005['Points:1'], cfs_si2_005['C'], linestyle='-',label='massTransferVoF, t = 05 s', color='#bf812d')
ax.plot(cfs_si2_010['Points:1'], cfs_si2_010['C'], linestyle='-',label='massTransferVoF, t = 10 s', color='#dfc27d')
ax.plot(cfs_si2_020['Points:1'], cfs_si2_020['C'], linestyle='-',label='massTransferVoF, t = 20 s', color='#c7eae5')
ax.plot(cfs_si2_030['Points:1'], cfs_si2_030['C'], linestyle='-',label='massTransferVoF, t = 30 s', color='#80cdc1')
ax.plot(cfs_si2_040['Points:1'], cfs_si2_040['C'], linestyle='-',label='massTransferVoF, t = 40 s', color='#35978f')
ax.plot(cfs_si2_100['Points:1'], cfs_si2_100['C'], linestyle='-',label='massTransferVoF, t = 100 s', color='#01665e')

plt.grid(True, which='both', linestyle='--', linewidth=0.5, alpha=0.7)

plt.xlabel('Distance y [m]')
plt.ylabel('Concentration C [mol/m]')
plt.legend()

fig.savefig("../figures/Nieves-Remacha_2015_SI-2.png", dpi=300)

plt.show()
```
