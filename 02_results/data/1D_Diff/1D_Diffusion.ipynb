# Stofftransport - 1D-Diffusion

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df_cfd_2e6 = pd.read_csv('1D_Diff_t2e6s.csv')
df_cfd_6e6 = pd.read_csv('1D_Diff_t6e6s.csv')
df_cfd_20e6 = pd.read_csv('1D_Diff_t20e6s.csv')
df_cfd_100e6 = pd.read_csv('1D_Diff_t50e6s.csv')

df_ref_phi_6e6 = pd.read_csv('1D_Diff_t6e6s_phi.csv', header=None, names=['x','y'])
df_ref_phi_2e6 = pd.read_csv('1D_Diff_t2e6s_phi.csv', header=None, names=['x','y'])
df_ref_phi_20e6 = pd.read_csv('1D_Diff_t20e6s_phi.csv', header=None, names=['x','y'])
df_ref_phi_100e6 = pd.read_csv('1D_Diff_t100e6s_phi.csv', header=None, names=['x','y'])
```

```python
plt.figure()
plt.grid(True, which='both', linestyle='--', linewidth=0.5, alpha=0.7)

plt.plot(df_ref_phi_2e6['x'], df_ref_phi_2e6['y'], label='Reference Solution, t = 2e6 s', color='#a6611a')
plt.plot(df_ref_phi_6e6['x'], df_ref_phi_6e6['y'], label='Reference Solution, t = 6e6 s', color='#8c510a')
plt.plot(df_ref_phi_20e6['x'], df_ref_phi_20e6['y'], label='Reference Solution, t = 2e7 s', color='#bf812d')
plt.plot(df_ref_phi_100e6['x'], df_ref_phi_100e6['y'], label='Reference Solution, t = 1e8 s', color='#dfc27d')
plt.plot(df_cfd_2e6['Points:0'], df_cfd_2e6['C'], 'o', label='massTransferVoF, t = 2e6 s', markersize=4, markeredgecolor='#01665e', markerfacecolor='None', linestyle='None')  # 'o' = Punkte
plt.plot(df_cfd_6e6['Points:0'], df_cfd_6e6['C'], 'o', label='massTransferVoF, t = 6e6 s', markersize=4, markeredgecolor='#35978f', markerfacecolor='None', linestyle='None', markevery=10)  # 'o' = Punkte
plt.plot(df_cfd_20e6['Points:0'], df_cfd_20e6['C'], 'o', label='massTransferVoF, t = 2e7 s', markersize=4, markeredgecolor='#80cdc1', markerfacecolor='None', linestyle='None', markevery=10)  # 'o' = Punkte
plt.plot(df_cfd_100e6['Points:0'], df_cfd_100e6['C'], 'o', label='massTransferVoF, t = 1e8 s', markersize=4, markeredgecolor='#c7eae5', markerfacecolor='None', linestyle='None', markevery=10)  # 'o' = Punkte

#plt.title('He=10')
#plt.title('a', loc='left')
plt.xlabel('Distance y [m]')
plt.ylabel('Concentration C [mol/m³]')
plt.legend(loc='upper left', prop={'size': 8})
plt.savefig("1D_Diff_comp.png", dpi=300)
plt.show()
```
